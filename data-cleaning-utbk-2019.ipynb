```python
# import useful libraries

import pandas as pd
import numpy as np
```

```python
# i make 3 different dataframes that i will merge later

# df1_nilai contains choice of university & major code and test scores

# df_2_univ contains university code and university name

# df3_prodi contains major code and name

df1_nilai = pd.read_excel('utbk-saintek.xlsx', sheet_name = 'nilai')
df2_univ = pd.read_excel('utbk-saintek.xlsx', sheet_name = 'univ')
df3_prodi = pd.read_excel('utbk-saintek.xlsx', sheet_name = 'prodi')
```

```python
# checking df1_nilai

df1_nilai.head()
```

```python
# checking df2_nilai

df2_univ.head()
```

```python
# checking df3_nilai

df3_prodi.head()
```

```python
# create new dataframe to merge df1_nilai with df2_univ based on KODE PTN or university code

df4 = df1_nilai.merge(df2_univ, on = 'KODE PTN', how='left')

df4.head(20)
```

```python
# now i create new dataframe to merge df4 with df3_prodi based on KODE PRODI or major code

df5 = df4.merge(df3_prodi, on = 'KODE PRODI', how='left')
```

```python
# checking df5

df5.head()
```

```python
# i basically rearrange the column order with the easiest way

df5 = df5[['KODE PRODI', 'NAMA PRODI', 'KODE PTN', 'PTN', 'id_user','score_bio','score_fis',
           'score_kim','score_kmb','score_kpu','score_kua','score_mat','score_ppu']]
```

```python
# checking df5 again

df5.head()
```

```python
# something seems odd, isn't it? Df5 dataframe contains so many duplicate row. Let's remove them
```

```python
df5 = df5.drop_duplicates()
```

```python
# check again

df5.head()

```

```python
# do you see that 'NaN' values? I'm getting suspicious here. Let's check the NaN values i get.

nan_df = df5[df5.isna().any(axis=1)]
nan_df.head()
```

```python
# well, looks like i got plenty of them. Let's say goodbye to NaN!

df5 = df5.dropna(how='any')
df5.dropna(how='any', inplace=True)

# checking df5
df5.head()
```

```python
# did they go already? let's check them one more time.

# basically if the output is True it means i still get some nulls, and vice versa.

print(df5.isnull().values.any())
```

```python
# False, so there's no null values left.

# checking df5 again

df5.head()
```

```python
# now, let's open a new excel file. This is the passing grade file

# passing grade will determine whether the participant will get accepted or not, based on their test score

# calling passing grade file with df6

df6 = pd.read_excel('pg.xlsx')
df6.head()
```

```python
# i create new data frame to merge passing grade file with df5 based on KODE PRODI or major code

df7_hasil_utbk = df6.merge(df5, on = 'KODE PRODI', how='left')
```

```python
# checking df7_hasil_utbk

df7_hasil_utbk.head()
```

```python
# the nulls are coming back! this has the same condition as i got before

# so i'm just going to straight up drop the nulls

df7_hasil_utbk = df7_hasil_utbk.dropna(how='any')
```

```python
# check df7_hasil_utbk again

df7_hasil_utbk.head()
```

```python
# check if i still have some nulls left

print(df7_hasil_utbk.isnull().values.any())
```

```python
# nice, now i want to check the column's datatypes because they just doesnt feel right

df7_hasil_utbk.info()

# KODE PTN (university code) is float??

# id_user (participant code) is float??
```

```python
# let's change them to integer or basic number

df7_hasil_utbk = df7_hasil_utbk.astype({"KODE PTN": int, "id_user":int})
```

```python
# check df7 again

df7_hasil_utbk.head()
```

```python
# the next task

# i want to calculate the total score for each participant. The score will determine whether or not they will get accepted, based on the passing grade

# total score = (bio + fis + kim + kmb + kpu + kua + mat + ppu) / 8

total_skor = df7_hasil_utbk["score_bio"] + df7_hasil_utbk["score_fis"] + df7_hasil_utbk["score_kim"] + df7_hasil_utbk["score_kmb"] + df7_hasil_utbk["score_kpu"] + df7_hasil_utbk["score_kua"] + df7_hasil_utbk["score_mat"] + df7_hasil_utbk["score_ppu"]

df7_hasil_utbk["TOTAL SKOR"] = total_skor / 8
```

```python
# check df7 again

df7_hasil_utbk.head()
```

```python
# nice

# next, i want to create new column that contains the result of their test with conditional syntax

# if TOTAL SKOR < PASSING GRADE, then NOT ACCEPTED (TIDAK LOLOS)

# if TOTAL SKOR >= PASSING GRADE, then ACCEPTED (LOLOS)

df7_hasil_utbk.loc[df7_hasil_utbk['TOTAL SKOR'] < df7_hasil_utbk['PASSING GRADE'], 'HASIL'] = 'TIDAK LOLOS' 
df7_hasil_utbk.loc[df7_hasil_utbk['TOTAL SKOR'] >= df7_hasil_utbk['PASSING GRADE'], 'HASIL'] = 'LOLOS'
```

```python
# check check

df7_hasil_utbk.head()
```

```python
# then, i save the final result to excel again
```

```python
df7_hasil_utbk.to_csv("hasil_utbk.csv", index=False)
```
